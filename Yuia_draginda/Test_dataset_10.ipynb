```python
import os
im_path = '/content/test10-1/train'
names = sorted(os.listdir(im_path))[:-1]
names
```

```python
import shutil
target_path = '/content/masks_Yulia_Draginda'

if os.path.exists(target_path):
    shutil.rmtree(target_path)
os.makedirs(target_path)
```

```python
im_count = 0
for k in range(len(names)):
  im_name = names[k]
  if im_name.endswith('.jpg'):
      im_n = im_name.split('.')[0][:-4]+'.jpg'
      shutil.copy(os.path.join(im_path, im_name), os.path.join(target_path, im_n))

  if im_name.endswith('.png'):
      im_n = im_name.split('.')[0][:-4]+'_mask'+'.png'
      shutil.copy(os.path.join(im_path, im_name), os.path.join(target_path, im_n))
  im_count+=1
```

```python
import matplotlib.pyplot as plt
count = 0
print("Визуализация раметки:")
for n in sorted(os.listdir(target_path)):
  if 'mask' in n:
    msk = plt.imread(target_path+'/'+n)
  else:
    im = plt.imread(target_path+'/'+n)
  count+=1
  if count%2==0:
    plt.imshow(im)
    plt.imshow(msk, alpha = 0.3)
    plt.axis('off')
    plt.show()
```
